## Check sales trend by months:

```python
sales_by_month=pd.pivot_table(sales, values='total_sales', index=['month'], columns=['year'], aggfunc=np.sum)
sales_by_month.reset_index(inplace=True)
sales_by_month
```

```python
plt.plot(sales_by_month['month'], sales_by_month['2019'])
```

```python
## Based on the data and chart, from Jan. to Nov. the sales was increasing, especially from May to Nov.,
## sales increasement was very fast. But there was a dramatic decreasement in Dec, there must be something 
## happened. The same situation happened in Dec., 2018 as well.The reason for this drop should be becasue 
## customers have to wait for several weeks to receive the orders, so they bought gifts in Nov or even Oct. 
## If we check the canceled orders in Dec. , 2019, there were 347 orders cancled. That might
## be one of the reasons why the sales dropped so big.One of the reason for those canceled orders
## might be the shipping was too long, or there were not enough stock. 
```

### Check order size per transaction

```python
sales_per_transaction = pd.pivot_table(sales, values=['total_sales','quantity'], index=['transaction_no','country'],   aggfunc=np.sum)
sales_per_transaction.reset_index(inplace=True)
sales_per_transaction
```

```python
## average order size by country
```

```python
ave_sales_per_transaction = pd.pivot_table(sales, values=['total_sales','quantity'], index=['country'],   aggfunc=np.mean)
ave_sales_per_transaction.reset_index(inplace=True)
round(ave_sales_per_transaction,0).sort_values(by=['quantity','total_sales'], ascending = False)
```

```python
round(ave_sales_per_transaction.describe(),0)
```

```python
top_ave_sales_per_transaction=ave_sales_per_transaction.loc[ ave_sales_per_transaction['quantity']>=19, :]
round(top_ave_sales_per_transaction,0).sort_values(by=['quantity'], ascending = False)
```

```python
## From the average datab, one interesting thing is that these countries which have top average order size are not those countries who
## have large sales. It means people in those top average size countries prefer to buy more products online. These countries actually
## are more profitable. Due to those countries have small total sales right now, they should be potential to sell more 
## if the company has the proper business development strateges.
```
